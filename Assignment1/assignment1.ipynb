```python
maze = {1: {'E'}, 
        2: {'N', 'W'}, 
        3: {'E', 'N'}, 
        4: {'N', 'W'}, 
        5: {'E', 'N'},
        6: {'S', 'W', 'E'}, 
        7: {'S', 'W'}, 
        8: {'S', 'N'}, 
        9: {'S', 'N'}, 
        10: {'N'},
        11: {'N', 'E'}, 
        12: {'S', 'W'}, 
        13: {'S', 'E'}, 
        14: {'S', 'W'}, 
        15: {'S', 'E'},
        16: {'W'}
        }
```

```python
def allowedMove(loc, move):
    # maze: dict of possible moves per node
    # loc: integer label of current node
    # move: ['W','S','E','N']
    return (move in maze[loc])
```

```python
def neighbour(maze, loc, move):
    # maze: dict of possible moves per node
    # loc: integer label of current node
    # move: ['W','S','E','N']
    if move == 'N':
        res = loc + 4
    if move == 'S':
        res = loc - 4
    if move == 'E':
        res = loc + 1
    if move == 'W':   
        res = loc - 1        
    if res > 0 and res < 17: return res
    return loc
```

```python
def openMazeDFS(maze, start, goal, M = 16):
    stack = []
    stack.append(start)
    path = []
    i = 0;
    while len(stack) > 0 :
        loc = stack.pop()
        path.append(loc)
        if loc == goal:
            print('Goal found')
            return path
        for move in ['W','S','E','N']:
            if allowedMove(loc, move):
                stack.append(neighbour(maze, loc, move))
        i = i + 1
        if i > M: 
            print("Error")
            return path
    print ("Goal not found")
```

```python
len(openMazeDFS(maze, 1, 15))
```

```python
def closedMazeDFS(maze, start, goal, order = ['N', 'E', 'S', 'W']):
    stack = []
    stack.append(start)
    visited = set()
    order.reverse()
    path = []
    while len(stack) > 0 :
        loc = stack.pop()
        visited.add(loc)
        path.append(loc)
        if loc == goal:
            print('Goal found')
            return path
        for move in order:
            if allowedMove(loc, move):
                temp = neighbour(maze, loc, move)
                if temp not in visited and temp not in stack:
                    stack.append(temp)
    print ("Goal not found")
    return path
```

```python
closedMazeDFS(maze, 1, 15)
```

```python
closedMazeDFS(maze, 1, 15, order = ['N','S','W','E'])
```

```python
def openMazeBFS(maze, start, goal):
    queue = []
    queue.append(start)
    path = []
    while len(queue) > 0 :
        loc = queue.pop(0)
        path.append(loc)
        if loc == goal:
            print('Goal found')
            return path
        for move in ['W','S','E','N']:
            if allowedMove(loc, move):
                queue.append(neighbour(maze, loc, move))
    print ("Goal not found")
    return path
```
